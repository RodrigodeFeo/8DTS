```python
# Obter as sequências de fechamento dos últimos 15 dias
X_train = df_petro_train.iloc[:, 5:].values  # Supondo que as colunas de sequência de fechamento começam da segunda coluna
X_test = df_preto_test.iloc[:, 5:].values

# Normalizar os dados de fechamento
X_train = (X_train - X_train.min(axis=1, keepdims=True)) / (X_train.max(axis=1, keepdims=True) - X_train.min(axis=1, keepdims=True))
X_test = (X_test - X_test.min(axis=1, keepdims=True)) / (X_test.max(axis=1, keepdims=True) - X_test.min(axis=1, keepdims=True))

# Obter os rótulos de compra/venda do DataFrame
y_train = df_petro_train['Label'].values
y_test = df_preto_test['Label'].values

# Converter os rótulos para o formato correto (-1 para vender, 1 para comprar)
y_train = np.where(y_train == -1, 0, 1)
y_test = np.where(y_test == -1, 0, 1)

# Redimensionar os dados para (samples, timesteps, features) para o modelo 1D CNN
X_train = np.expand_dims(X_train, axis=2)
X_test = np.expand_dims(X_test, axis=2)

```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(15, 1)),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(50, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()

```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
#1. Acurácia no Conjunto de Teste
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
#2. Matriz de Confusão, Precisão e Recall
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# Fazer previsões no conjunto de teste
y_pred = (model.predict(X_test) > 0.5).astype("int32")

# Calcular a matriz de confusão
conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusão:")
print(conf_matrix)

# Calcular a precisão
precision = precision_score(y_test, y_pred)
print(f"Precisão: {precision:.4f}")

# Calcular o recall
recall = recall_score(y_test, y_pred)
print(f"Recall: {recall:.4f}")


import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Fazer previsões no conjunto de teste
y_pred = (model.predict(X_test) > 0.5).astype("int32")

# Calcular a matriz de confusão
conf_matrix = confusion_matrix(y_test, y_pred)

# Plotar a matriz de confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Fazer previsões no conjunto de dados de preços históricos
# Supondo que df_prices contém colunas 'date' e 'close'
# Vamos gerar sequências de 15 dias de preços de fechamento para alimentar o modelo

df_prices = df_preto_test.copy()

window_size = 15
sequences = []
for i in range(len(df_prices) - window_size):
    sequences.append(df_prices['Close'].values[i:i + window_size])

sequences = np.array(sequences)
sequences = (sequences - sequences.min(axis=1, keepdims=True)) / (sequences.max(axis=1, keepdims=True) - sequences.min(axis=1, keepdims=True))
sequences = np.expand_dims(sequences, axis=2)

# Fazer previsões usando o modelo treinado
predictions = (model.predict(sequences) > 0.5).astype("int32")

# Calcular os retornos financeiros simulados
initial_balance = 10000  # Saldo inicial em dinheiro
balance = initial_balance
shares = 0  # Quantidade inicial de ações

for i in range(len(predictions)):
    if predictions[i] == 1 and balance >= df_prices['Close'].values[i + window_size]:
        # Comprar uma ação
        shares += 1
        balance -= df_prices['Close'].values[i + window_size]
    elif predictions[i] == 0 and shares > 0:
        # Vender uma ação
        shares -= 1
        balance += df_prices['Close'].values[i + window_size]

# Calcular o valor final do portfólio
final_balance = balance + shares * df_prices['Close'].values[-1]

# Calcular o retorno acumulado
cumulative_return = (final_balance - initial_balance) / initial_balance
print(f"Retorno acumulado do backtest: {cumulative_return:.4f}")
```

```python

import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_score, recall_score
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Definindo os caminhos para os diretórios de treino e teste
train_dir = 'treino'
test_dir = 'teste'

# Usando ImageDataGenerator para carregar as imagens
train_datagen = ImageDataGenerator(rescale=0.1/255.0)
test_datagen = ImageDataGenerator(rescale=0.1/255.0)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)

# Mostrar as classes do gerador
print(train_generator.class_indices)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model_cnn = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model_cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_cnn.summary()

# Treinar o modelo
history_cnn = model_cnn.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=20,
    validation_data=test_generator,
    validation_steps=test_generator.samples // test_generator.batch_size
)
```

```python
# Avaliar o modelo no conjunto de teste
loss_cnn, accuracy_cnn = model_cnn.evaluate(test_generator)
print(f'Acurácia no conjunto de teste: {accuracy_cnn:.4f}')

# Fazer previsões no conjunto de teste
test_generator.reset()
y_pred_cnn = (model_cnn.predict(test_generator) > 0.5).astype("int32").flatten()

# Obter os rótulos verdadeiros
y_test_cnn = test_generator.classes

# Calcular a matriz de confusão
conf_matrix_cnn = confusion_matrix(y_test_cnn, y_pred_cnn)
print("Matriz de Confusão:")
print(conf_matrix_cnn)

# Plotar a matriz de confusão
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_cnn, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix (2D CNN)')
plt.show()

# Calcular a precisão
precision_cnn = precision_score(y_test_cnn, y_pred_cnn)
print(f"Precisão: {precision_cnn:.4f}")

# Calcular o recall
recall_cnn = recall_score(y_test_cnn, y_pred_cnn)
print(f"Recall: {recall_cnn:.4f}")
```
