## Respostas abaixo:

Exercicio 1 - Preencher NA com valores Arbritários

```python
df_arbri = df
```

```python
df_arbri.fillna(-100000000,inplace = True)
```

```python
df_arbri
```

Exercicio 2 - Preencher NA com valores usando mediana

```python
df_mediana = df
```

```python
df_mediana.fillna(df_mediana.median(),inplace = True)
```

```python
df_mediana
```

Exercicio 3 - Preencher NA com valores usando média

```python
df_media = df
```

```python
df_media.fillna(df_media.mean(),inplace = True)
```

```python
df_media
```

Exercicio 4 - Preencher NA com valores usando End Of Tail

```python
df_endoftail = df
```

```python
df_endoftail.fillna(df_endoftail.ffill().tail(1).iloc[0],inplace = True)
```

```python
df_endoftail
```

Exercicio 5 - Preencher NA com o KNN Inputer

```python
df_knn = df
```

```python
knn_input = KNNImputer(n_neighbors=1)
knn_input.fit_transform(df_knn)
```
